# (1) PyeIQ 檢視 - SWITCH_IMAGE DEMO

```python
#路線設置
import os
#HOME = os.environ['HOME']
BASE_DIR = os.path.join('home\\root', ".cache", "eiq");
bin_dir  = os.path.join(BASE_DIR, "bin")
media_dir= os.path.join(BASE_DIR, "media")
model_dir= os.path.join(BASE_DIR, "model")
model_dir
```

# (2)  i.MX8 PyeIQ NPU Using Way
### Using Method

```python
#example for amazon's khadas
#https://docs.khadas.com/vim3/HowToUseNpu.html
```

```python
accel = 1
image = 'weilly.jpg' #測試影像
RUN_LABEL_IMAGE = "VSI_NN_LOG_LEVEL=0 {0} -m {1} -t 1 -i {2} -l {3} -a {4} -v 0 -c 100" #使用方式 - 請於 i.MX8MP 上運行此行
run_li = RUN_LABEL_IMAGE.format(binary, model, image, labels, accel)
run_li
```

```python
#NPU
VSI_NN_LOG_LEVEL=0 /home/root/.cache/eiq/eIQSwitchLabelImage/bin/label_image -m /home/root/.cache/eiq/eIQSwitchLabelImage/model/mobilenet_v1_1.0_224_quant.tflite -t 1 -i /home/root/.cache/eiq/eIQSwitchLabelImage/media/Scuba_Dive.bmp -l /home/root/.cache/eiq/eIQSwitchLabelImage/model/labels.txt -a 1 -v 0 -c 100

#CPU
VSI_NN_LOG_LEVEL=0 /home/root/.cache/eiq/eIQSwitchLabelImage/bin/label_image -m /home/root/.cache/eiq/eIQSwitchLabelImage/model/mobilenet_v1_1.0_224_quant.tflite -t 1 -i /home/root/.cache/eiq/eIQSwitchLabelImage/media/Scuba_Dive.bmp -l /home/root/.cache/eiq/eIQSwitchLabelImage/model/labels.txt -a 0 -v 0 -c 100
```
